```python
#Creating a dataframe with Nan values
df=pd.DataFrame([[1,6.5,3],[1,np.nan,np.nan],[np.nan,np.nan,np.nan],[np.nan,6.5,3]])
df
```

```python
df_no_missing=df.dropna() #Here we are droping all values which have nan values
df_no_missing
```

```python
df_cleaned=df.dropna(how='all')#Its mean if there is any row which have all nan values this formula will drop the rows
df_cleaned
```

```python
df['final_score']=np.nan #Create a new Column full of missing values
print(df)
```

```python
df.dropna(axis=1,how='all') #Drop Column if they only contain Missing Values
```

```python
#Fillna()
df.fillna(0)
df['final_score'].fillna(df['final_score'].mean(),inplace=True)
df
```

```python
#using grouby
df.groupby('final_score').count()
```

```python
df.fillna(method='bfill')
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
titanic_train=pd.read_csv('https://raw.githubusercontent.com/training-ml/Files/main/titanic_train.csv')
```

```python
titanic_train.head()
```

```python
titanic_train['Survived'].unique()
```

```python
new_survived=pd.Categorical(titanic_train['Survived'])
new_survived=new_survived.rename_categories(['Died','Survived'])
new_survived.describe()
```

```python
titanic_train['Pclass'].unique()
```

```python
new_place=pd.Categorical(titanic_train['Pclass'],ordered=True)
new_place=new_place.rename_categories(['class1','class2','class3'])
new_place.describe()
```

```python
titanic_train['Pclass']=new_place #Reassign New Name to pclass
```

```python
age_missing=titanic_train[titanic_train.Age.isin(['Nan'])] ##Find all the passengers who's age is Missing.
age_missing
```

```python
#fetch rows based on list of numbers like titanic_train[titanic_train.Age.isin([15])]
#you cannot pass condtition inside 'isin()'
```

```python
titanic_train.isna().sum()
```

```python
#Find only those passengers who's age is present in the dataset(equal to not in
age_present=titanic_train[~titanic_train.Age.isin(['Nan'])]
age_present
```
