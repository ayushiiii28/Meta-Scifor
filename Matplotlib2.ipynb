```python
#Boxplot
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.DataFrame([10,12,13,1,4,16,50])
plt.figure(figsize=(5,6))
ax = sns.boxplot(data=df)
plt.yticks(range(5,60,5)) #range on y axis
plt.show()

```

```python
#Another way
df.plot(kind='box',figsize=(8,5),yticks=range(0,55,5))
```

```python
#To remove outliers, we use IQR
#Inter Quantile Range
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
ser = pd.Series([15,22,24,25,30,23,34,43])
q1 = ser.quantile(0.25)
print('Q1:',q1)
q3 = ser.quantile(0.75)
print('Q3: ',q3)
iqr = q3 - q1
print('IQR: ',iqr)
```

```python
#Higher Side
#H = q3 + (1.5*iqr)
#Lower Side
#L = q1 - (1.5*iqr)
```

```python
upside = q3 + (1.5*iqr)
print('Upper Side: ',upside)
downside = q1 - (1.5*iqr)
print('Lower Side: ',downside)
```
